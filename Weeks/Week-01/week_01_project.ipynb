# Week 01 Project: Python Quiz
This notebook explains the quiz program one step at a time for a beginner student.


## 1. Import the random module
The `random` module helps us mix up the questions so the quiz is different every time.


```python
import random
```

## 2. Store the quiz questions
We keep all questions in a list. Each question is a dictionary with the question text, answer choices, and the correct answer.


```python
questions = [
    {
        'question': 'What is the capital of Kenya?',
        'choices': {
            'A': 'Nairobi',
            'B': 'Mombasa',
            'C': 'Kisumu',
            'D': 'Nakuru'
        },
        'answer': 'A'
    },
    {
        'question': 'Which keyword is used to create a loop in Python?',
        'choices': {
            'A': 'repeat',
            'B': 'loop',
            'C': 'for',
            'D': 'iterate'
        },
        'answer': 'C'
    },
    {
        'question': 'What is 5 + 7?',
        'choices': {
            'A': '10',
            'B': '12',
            'C': '13',
            'D': '14'
        },
        'answer': 'B'
    },
    {
        'question': 'Which data type stores True or False?',
        'choices': {
            'A': 'String',
            'B': 'Integer',
            'C': 'Boolean',
            'D': 'Float'
        },
        'answer': 'C'
    }
]
```

## 3. Shuffle the questions
We use `random.shuffle` to change the question order. That makes the quiz different each time.


```python
random.shuffle(questions)
```

## 4. Start the score at 0
The variable `score` keeps track of how many questions the user answers correctly.


```python
score = 0
```

## 5. Show the quiz title
This prints a title so the user knows the program is a quiz.


```python
print('=== PYTHON QUIZ ===')
```

## 6. Ask each question
We use a `for` loop to go through every question in the list.


```python
for q in questions:
    print('
' + q['question'])
    for key, value in q['choices'].items():
        print(f'{key}. {value}')

    while True:
        try:
            user_answer = input('Enter your answer (A/B/C/D): ').upper()
            if user_answer not in ['A', 'B', 'C', 'D']:
                raise ValueError
            break
        except ValueError:
            print('Invalid input! Please enter A, B, C, or D.')

    if user_answer == q['answer']:
        print('Correct!')
        score += 1
    else:
        print(f'Wrong! Correct answer is {q['answer']}')
```

## 7. Score and feedback
After the loop finishes, we show the score and give feedback based on the percentage.


```python
print('
=== QUIZ COMPLETE ===')
print(f'Your score: {score}/{len(questions)}')
percentage = (score / len(questions)) * 100
if percentage >= 80:
    print('Excellent!')
elif percentage >= 50:
    print('Good!')
else:
    print('Try Again!')
```

## 8. What this program teaches
This quiz program uses: variables, lists, dictionaries, loops, input, conditionals, and error handling.
