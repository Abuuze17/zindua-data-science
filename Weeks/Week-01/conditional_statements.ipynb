# Conditional Statements Practice
This notebook explains how `if`, `elif`, and `else` work in a simple calculator.


## Step 1: Get two numbers from the user
We use `input()` twice to read the first number and the second number. Then we convert each value to a number with `float()` so we can do math.


```python
num1 = float(input('Enter first number: '))
num2 = float(input('Enter second number: '))
```

## Step 2: Ask which operation to do
The user can choose `+`, `-`, `*`, or `/`. We store that choice in the variable `op`.


```python
op = input('Choose operation (+, -, *, /): ')
```

## Step 3: Use `if`, `elif`, and `else` to choose the answer
Python checks each condition in order. The first true condition runs, and the rest are skipped. If none are true, the `else` block runs.


```python
if op == '+':
    result = num1 + num2
elif op == '-':
    result = num1 - num2
elif op == '*':
    result = num1 * num2
elif op == '/':
    if num2 != 0:
        result = num1 / num2
    else:
        result = 'Error: Division by zero'
else:
    result = 'Invalid operation'
```

## Step 4: Print the result
We show the value stored in `result`. If the operation was wrong, the program tells the user about it.


```python
print('Result:', result)
```

## What we learned
- `if` checks one condition
- `elif` checks another condition if the first was false
- `else` runs when no condition is true
- nested `if` can check more details, like dividing by zero
