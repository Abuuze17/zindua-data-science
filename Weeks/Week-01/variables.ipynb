# Variables Practice
This notebook explains how variables work in Python with a small age calculator.


## Step 1: Ask for the user age
We use `input()` to get text from the user. The value from `input()` is always a string, even if the user types numbers.


```python
age_str = input('Enter your age: ')
```

## Step 2: Convert the text into a number
A variable can store text or a number. We need a number to do math, so we convert the string using `int()`.


```python
age = int(age_str)
```

## Step 3: Add 5 years to the age
The variable `age` now holds a number. We can add 5 to it to find the age in 5 years.


```python
future_age = age + 5
```

## Step 4: Show the result
We use `print()` to show the final message. This uses the value inside the `future_age` variable.


```python
print(f'In 5 years, you will be {future_age} years old.')
```

## What we learned
- `input()` gets text from the user
- variables can store text or numbers
- `int()` changes text into an integer number
- `print()` shows the result to the user
